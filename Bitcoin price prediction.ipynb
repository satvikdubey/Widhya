```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading and reading the dataset
df=pd.read_csv("https://raw.githubusercontent.com/WidhyaOrg/datasets/master/bitcoin_dataset.csv")
df.head()
```

```python
#Cleaning Data
df.isnull().sum()
```

```python
#filling all null values with the mean values
df["btc_total_bitcoins"].fillna(df["btc_total_bitcoins"].mean(skipna=True),inplace=True)
df["btc_trade_volume"].fillna(df["btc_trade_volume"].mean(skipna=True),inplace=True)
df["btc_blocks_size"].fillna(df["btc_blocks_size"].mean(skipna=True),inplace=True)
df["btc_median_confirmation_time"].fillna(df["btc_median_confirmation_time"].mean(skipna=True),inplace=True)
df["btc_difficulty"].fillna(df["btc_difficulty"].mean(skipna=True),inplace=True)
df["btc_transaction_fees"].fillna(df["btc_transaction_fees"].mean(skipna=True),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
x=df[['btc_market_cap','btc_n_transactions','btc_miners_revenue',
          'btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']]
y=df['btc_market_price']
```

```python
#Split data into train and test
X_train,x_test,Y_train,y_test = train_test_split(x,y,test_size=0.25, random_state= 1)
```

```python
#Prediction using Linear regression
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = lr.predict(x_test)
pred = pd.DataFrame({"Actual Price":y_test ,"Predicted price":Y_pred})
pred
```

```python
mse= mean_squared_error(y_test,Y_pred)
mse
```
