```python
import numpy as np
import pandas as pd 
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/WidhyaOrg/datasets/master/instagram_reach.csv")
df
```

```python
for i in range(len(df)):
    df.loc[i,"Time since posted"] = int(df.loc[i,"Time since posted"].replace("hours","").strip())
    
print("done")
```

```python
df
```

```python
df=df.drop(["USERNAME","Caption","Hashtags"],axis=1)
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred= model.predict(X_test)
y_pred
```

```python
mean_squared_error(y_test, y_pred)
```
